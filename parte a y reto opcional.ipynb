DASHBOARD COMPLETO

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from google.colab import files

%matplotlib inline

print("=" * 60)
print("INICIANDO DASHBOARD COMPLETO")
print("=" * 60)

# Cargar datos
df = pd.read_csv('sales_data_2024.csv')
print(f"Datos cargados: {len(df)} registros")

# CREAR LA FIGURA PRINCIPAL UNA SOLA VEZ
fig = plt.figure(figsize=(18, 12))
fig.suptitle('DASHBOARD DE ANALISIS DE VENTAS 2024 - FESC',
             fontsize=20, fontweight='bold', color='#C5282F')

# ============================================================================
# GRAFICA 1: VENTAS POR PRODUCTO (HORIZONTAL PARA MEJOR LECTURA)
# ============================================================================
ax1 = plt.subplot(2, 2, 1)
ventas_por_producto = df.groupby('Product')['Sales'].sum().sort_values(ascending=True)

y_pos = range(len(ventas_por_producto))
bars = ax1.barh(y_pos, ventas_por_producto.values, color='#2E86AB', alpha=0.8)

ax1.set_yticks(y_pos)
ax1.set_yticklabels(ventas_por_producto.index, fontsize=10)
ax1.set_title('Ventas Totales por Producto', fontweight='bold', fontsize=12)
ax1.set_xlabel('Ventas ($)', fontweight='bold')
ax1.grid(True, alpha=0.3)

# Agregar valores a las barras
for i, bar in enumerate(bars):
    width = bar.get_width()
    ax1.text(width + 100000, bar.get_y() + bar.get_height()/2.,
             f'${width/1000000:.1f}M',
             ha='left', va='center', fontsize=9, fontweight='bold')

print("Grafica 1: Ventas por Producto - COMPLETADA")

# ============================================================================
# GRAFICA 2: DISTRIBUCION POR REGION
# ============================================================================
ax2 = plt.subplot(2, 2, 2)
ventas_por_region = df.groupby('Region')['Sales'].sum()

colores = ['#C5282F', '#0077BE', '#28A745', '#FFC107', '#6C757D']
ax2.pie(ventas_por_region.values,
        labels=ventas_por_region.index,
        colors=colores,
        autopct='%1.1f%%',
        startangle=90)
ax2.set_title('Distribucion de Ventas por Region', fontweight='bold', fontsize=12)

print("Grafica 2: Distribucion por Region - COMPLETADA")

# ============================================================================
# GRAFICA 3: TENDENCIA DE VENTAS
# ============================================================================
ax3 = plt.subplot(2, 2, 3)
ventas_diarias = df.groupby('Date')['Sales'].sum()

ax3.plot(ventas_diarias.index, ventas_diarias.values,
         color='#28A745', linewidth=2.5, marker='o', markersize=4)
ax3.set_title('Tendencia de Ventas Diarias', fontweight='bold', fontsize=12)
ax3.set_xlabel('Fecha')
ax3.set_ylabel('Ventas ($)')
ax3.grid(True, alpha=0.3)
ax3.tick_params(axis='x', rotation=45)

print("Grafica 3: Tendencia de Ventas - COMPLETADA")

# ============================================================================
# GRAFICA 4: TOP 5 PRODUCTOS
# ============================================================================
ax4 = plt.subplot(2, 2, 4)
top_5_productos = ventas_por_producto.nlargest(5)

y_pos_top = range(len(top_5_productos))
bars_top = ax4.barh(y_pos_top, top_5_productos.values, color='#DC3545', alpha=0.8)
ax4.set_yticks(y_pos_top)
ax4.set_yticklabels(top_5_productos.index, fontsize=10)
ax4.set_title('Top 5 Productos Mas Vendidos', fontweight='bold', fontsize=12)
ax4.set_xlabel('Ventas ($)')
ax4.grid(True, alpha=0.3)

# Agregar valores al top 5
for i, bar in enumerate(bars_top):
    width = bar.get_width()
    ax4.text(width + 100000, bar.get_y() + bar.get_height()/2.,
             f'${width/1000000:.1f}M',
             ha='left', va='center', fontsize=9, fontweight='bold')

print("Grafica 4: Top 5 Productos - COMPLETADA")

# ============================================================================
# GUARDAR Y MOSTRAR
# ============================================================================
plt.tight_layout()

# GUARDAR PRIMERO
plt.savefig('mi_dashboard_ventas.png', dpi=150, bbox_inches='tight', facecolor='white')
print("Dashboard guardado como 'mi_dashboard_ventas.png'")

# MOSTRAR DESPUES
plt.show()

# DESCARGAR
files.download('mi_dashboard_ventas.png')
print("Archivo descargado a tu computadora")

print("=" * 60)
print("DASHBOARD COMPLETADO EXITOSAMENTE")
print("=" * 60)
```

```python
print("=" * 60)
print("RETO EXTRA: FUNCIONALIDADES AVANZADAS")
print("=" * 60)

import matplotlib.pyplot as plt
import pandas as pd
from datetime import datetime

# 1. AGREGAR QUINTA GRAFICA - Ventas por Categoria
print("1. Creando quinta grafica: Ventas por Categoria")

fig_extra = plt.figure(figsize=(15, 10))

# Grafica 5: Ventas por Categoria
ax5 = plt.subplot(2, 3, 1)
ventas_por_categoria = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)

bars_cat = ax5.bar(ventas_por_categoria.index, ventas_por_categoria.values,
                   color=['#FF6B6B', '#4ECDC4', '#45B7D1'], alpha=0.8)
ax5.set_title('Ventas por Categoria', fontweight='bold')
ax5.set_ylabel('Ventas ($)')
ax5.tick_params(axis='x', rotation=45)

# Agregar valores
for bar in bars_cat:
    height = bar.get_height()
    ax5.text(bar.get_x() + bar.get_width()/2., height + 100000,
             f'${height/1000000:.1f}M', ha='center', va='bottom', fontsize=9)

# 2. FILTRO POR FECHAS
print("2. Implementando filtro por fechas")

# Convertir columna Date a datetime
df['Date'] = pd.to_datetime(df['Date'])

# Filtrar por primer trimestre (Enero-Marzo)
fecha_inicio = '2024-01-01'
fecha_fin = '2024-03-31'
df_filtrado = df[(df['Date'] >= fecha_inicio) & (df['Date'] <= fecha_fin)]

# Grafica 6: Ventas del primer trimestre
ax6 = plt.subplot(2, 3, 2)
ventas_trimestre = df_filtrado.groupby('Product')['Sales'].sum().nlargest(5)

bars_trim = ax6.bar(range(len(ventas_trimestre)), ventas_trimestre.values,
                    color='#95E1D3', alpha=0.8)
ax6.set_title('Top 5 Productos - Primer Trimestre', fontweight='bold')
ax6.set_ylabel('Ventas ($)')
ax6.set_xticks(range(len(ventas_trimestre)))
ax6.set_xticklabels(ventas_trimestre.index, rotation=45, ha='right')

# 3. METRICAS RESUMEN
print("3. Calculando metricas de resumen")

ax7 = plt.subplot(2, 3, 3)
ax7.axis('off')  # Ocultar ejes para texto

# Calcular metricas
total_ventas = df['Sales'].sum()
venta_promedio = df['Sales'].mean()
total_productos = df['Product'].nunique()
region_top = df.groupby('Region')['Sales'].sum().idxmax()

texto_metricas = f"""
METRICAS PRINCIPALES:
-------------------
Ventas Totales: ${total_ventas:,.0f}
Venta Promedio: ${venta_promedio:,.0f}
Total Productos: {total_productos}
Region Lider: {region_top}
Total Transacciones: {len(df)}
Periodo Analizado: {df['Date'].min().strftime('%d/%m/%Y')} - {df['Date'].max().strftime('%d/%m/%Y')}
"""

ax7.text(0.1, 0.9, texto_metricas, fontsize=12, fontfamily='monospace',
         verticalalignment='top', linespacing=1.5)

# 4. GRAFICA DE METODOS DE PAGO
print("4. Analizando metodos de pago")

ax8 = plt.subplot(2, 3, 4)
ventas_por_pago = df.groupby('Payment_Method')['Sales'].sum()

ax8.pie(ventas_por_pago.values, labels=ventas_por_pago.index,
        autopct='%1.1f%%', startangle=90, colors=['#FFD93D', '#6BCF7F', '#4D96FF'])
ax8.set_title('Ventas por Metodo de Pago', fontweight='bold')

# 5. TENDENCIA MENSUAL
print("5. Creando tendencia mensual")

ax9 = plt.subplot(2, 3, 5)
df['Mes'] = df['Date'].dt.month
ventas_mensuales = df.groupby('Mes')['Sales'].sum()

meses = ['Ene', 'Feb', 'Mar', 'Abr']
ax9.plot(meses[:len(ventas_mensuales)], ventas_mensuales.values,
         marker='o', linewidth=2, color='#9B5DE5')
ax9.set_title('Tendencia de Ventas Mensual', fontweight='bold')
ax9.set_ylabel('Ventas ($)')
ax9.grid(True, alpha=0.3)

# 6. EXPORTAR A PDF
print("6. Exportando a PDF y PNG")

plt.tight_layout()

# Guardar en multiple formatos
plt.savefig('dashboard_avanzado.png', dpi=150, bbox_inches='tight')
plt.savefig('dashboard_avanzado.pdf', bbox_inches='tight')

# Mostrar resultados
plt.show()

# Descargar archivos
files.download('dashboard_avanzado.png')
files.download('dashboard_avanzado.pdf')

print("=" * 60)
print("RETO EXTRA COMPLETADO")
print("Se generaron 6 graficas adicionales")
print("Formatos exportados: PNG y PDF")
print("=" * 60)

# Mostrar resumen de filtros aplicados
print("\nRESUMEN DE FILTROS APLICADOS:")
print(f"Primer trimestre: {fecha_inicio} a {fecha_fin}")
print(f"Registros en filtro: {len(df_filtrado)} de {len(df)} totales")
print(f"Ventas del trimestre: ${df_filtrado['Sales'].sum():,.0f}")
```
